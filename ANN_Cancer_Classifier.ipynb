## Importing the libraries

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

## Importing the dataset

```python
path = "your dataset's path"
df = pd.read_csv(path)
df.head()
```

## Check missing Data

```python
missing_data = df.isnull()
missing_data.head()

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
    
#no missing data
```

## Checking the correlation

```python
corr = df.corr()
corr = corr["target"]
corr
```

## Dropp low correlation columns

```python
dataset = df.drop(['mean fractal dimension' , 'texture error' , 'smoothness error' , 'symmetry error' , 'fractal dimension error'] , axis=1)
dataset
```

## 1's VS 0's

```python
dataset['target'].value_counts()
```

## Splitting input & output

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

## Building ANN Model

```python
ann = tf.keras.models.Sequential()

#Input Layer
ann.add(tf.keras.layers.Dense(units=25, activation='relu')) 

#Hidden Layers
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))

#Output Layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# Training the Model

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 50)
```

## Predicting the Test set results and comparing them

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
